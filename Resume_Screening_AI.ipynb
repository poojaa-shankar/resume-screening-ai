```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Step 1: Define resumes
resumes = [
    "Python NLP automation experience",
    "SQL Power BI data analyst",
    "AI software engineer with Python and NLP"
]

# Step 2: Define job description
job_description = """
Looking for an AI Software Engineer with strong Python and NLP skills
"""

# Step 3: Combine all text
documents = resumes + [job_description]

# Step 4: Convert text to numerical vectors using TF-IDF
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(documents)

# Step 5: Calculate cosine similarity
similarity_scores = cosine_similarity(tfidf_matrix[:-1], tfidf_matrix[-1])

# Step 6: Rank resumes
print("Resume Ranking:\n")
for index, score in sorted(
    enumerate(similarity_scores),
    key=lambda x: x[1],
    reverse=True
):
    print(f"Resume {index + 1} similarity score: {score[0]:.2f}")
```
